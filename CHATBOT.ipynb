```python
import random
import re

class chatbot:
    negative_responses = ("no", "nah", "naw", "not a chance", "sorry")
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")

    random_questions = (
        "Why are you here?",
        "Are there many humans like you?",
        "What do you consume for sustenance?",
        "Is there intelligent life on this planet?",
        "Does Earth have a leader?",
        "What planets have you visited?",
        "What technology do you have on this planet?"
    )

    def __init__(self):
        self.patterns = {
            "earth": r'.*earth.*',
            "what_help": r'what\scan\sdo.*',
            "advise": r'.*advise.*',
        }
    def greet(self):
        self.name = input("Hi, I'm ChatBot. What's your name?\n")
        if self.name.lower() in self.negative_responses:
            print("Sorry, you're not providing your name.")
            return
        reply = input(f"Hi {self.name}, how are you?\n")
        self.chat()
            
    def make_exit(self,reply):
        if reply in self.exit_commands:
            print("done, bye!")
            return True
            
    def chat(self):
        reply = input(random.choice(self.random_questions)).lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply))
            if reply in self.negative_responses:
                print(" thinnaga answer cheyyi guru")
            
    def match_reply(self,reply):
        for key,pattern in self.patterns.items():
            found_match = re.match(pattern,reply)
            if found_match and key == "earth":
                return self.earth()
            elif found_match and key == "what_help":
                return self.what_help()
            elif found_match and key == "advise":
                return self.advice()
        if not found_match:
            return self.default()
    

    def earth(self):
        responses = [
            "Earth is like a large language model meta AI 3",
            "Earth is like your heart; it can hold anything, like you are holding girls in your heart."
        ]
        return random.choice(responses)

    def advice(self):
        return "BE GOOD, DO GOOD... UR RAJ"

    def what_help(self):
        return "Anything that you want!"

    def default(self):
        responses = [
            "Please tell me more.",
            "Tell me more!",
            "Why do you say that?",
            "I see. Can you elaborate?",
            "Interesting. Can you tell me more?",
            "I see. How do you think?",
            "Why?",
            "How do you think I feel when you say that?"
        ]
        return random.choice(responses)

ChatBot = chatbot()
ChatBot.greet()
```

```python
import random
import re

# Define responses for different patterns
responses = {
    'greeting': ['Hello!', 'Hi there!', 'Hey!', 'Greetings!'],
    'goodbye': ['Goodbye!', 'Bye!', 'See you later!', 'Take care!'],
    'thanks': ['You\'re welcome!', 'No problem.', 'My pleasure!'],
    'default': ['Sorry, I didn\'t understand that.', 'Could you please repeat that?']
}

# Define patterns and corresponding responses
patterns = {
    'greeting': re.compile(r'hello|hi|hey|good morning|good afternoon', re.IGNORECASE),
    'goodbye': re.compile(r'bye|goodbye|see you|take care', re.IGNORECASE),
    'thanks': re.compile(r'thank you|thanks|appreciate', re.IGNORECASE)
}

def get_response(user_input):
    user_input = user_input.lower()
    for key,pattern in patterns.items():
        if re.search(pattern, user_input):
            return random.choice(responses[key])
    return random.choice(responses['default'])

# Main loop for chatting

print("Bot: Hello! How can I help you today?")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Bot: Goodbye!")
        break
    else:
        response = get_response(user_input)
        print("Bot:", response)
```

```python
import nltk
from nltk.chat.util import Chat, reflections

# Define pairs of patterns and responses
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How can I help you today?",]
    ],
    [
        r"hi|hey|hello",["Hello!", "Hey there!",]
    ],
    [
        r"what is your name?",
        ["I am a chatbot created for a data science project.",]
    ],
    [
        r"how are you?",
        ["I'm just a program, but I'm here to help you!",]
    ],
    [
        r"quit",
        ["Goodbye!", "It was nice talking to you. See you soon!"]
    ],
]

# Default responses
default_responses = ["Sorry, I don't understand that.", "Can you please rephrase?"]

# Create a chatbot instance
chat = Chat(pairs, reflections)

def chatbot():
    print("Hi! I'm a chatbot. Type 'quit' to exit.")
    while True:
        user_input = input("You: ").lower()
        if user_input in ["quit", "exit"]:
            print("Goodbye!")
            break
        response = chat.respond(user_input)
        if response:
            print("Bot:", response)
        else:
            print("Bot:", default_responses[0])

if __name__ == "__main__":
    chatbot()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
import numpy as np
import random

nltk.download('punkt')
nltk.download('wordnet')
```

```python
training_data = [
    {"intent": "greeting", "patterns": ["Hi", "Hello", "How are you", "Good day"]},
    {"intent": "goodbye", "patterns": ["Bye", "See you later", "Goodbye"]},
    {"intent": "thanks", "patterns": ["Thanks", "Thank you", "That's helpful"]},
    {"intent": "noanswer", "patterns": ["None of these", "I don't know"]},
]

responses = {
    "greeting": ["Hello!", "Hi there!", "Greetings!", "How can I help you?"],
    "goodbye": ["Goodbye!", "See you later!", "Take care!"],
    "thanks": ["You're welcome!", "Happy to help!", "No problem!"],
    "noanswer": ["Sorry, I didn't get that.", "Can you please clarify?"],
}
```

```python
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens]
    return ' '.join(tokens)

training_sentences = []
training_labels = []

for data in training_data:
    for pattern in data["patterns"]:
        training_sentences.append(preprocess(pattern))
        training_labels.append(data["intent"])

labels = list(set(training_labels))
```

```python
model = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('classifier', MultinomialNB()),
])

model.fit(training_sentences, training_labels)
```

```python
def chatbot_response(text):
    processed_text = preprocess(text)
    predicted_label = model.predict([processed_text])[0]
    response = random.choice(responses[predicted_label])
    return response

# Test the chatbot
while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Chatbot: Goodbye!")
        break
    
    print(f"Chatbot: {chatbot_response(user_input)}")
```
